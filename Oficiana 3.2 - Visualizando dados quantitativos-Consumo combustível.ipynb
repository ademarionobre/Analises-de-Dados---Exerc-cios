## Consumo de combustível

Nesta oficina, você terá a oportunidade de trabalhar com um conjunto de dados com informações sobre o consumo de galão de combustível por milha (miles per galon, mpg), número de cilindros, cilindrada, cavalos de potência, peso, aceleração, ano do modelo, país de origem e nome do carro. Esse conjunto de dados chama-se “mpg” e é bastante famoso em estatística, sendo disponibilizado, diretamente, pela biblioteca seaborn sem a necessidade de baixar os dados para a sua máquina

```python
# importando bibliotecas e estilo de visualização.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
# dados direto do seaborn.
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
mpg.describe()
```

#### Gerar uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível diferenciar o país de origem (‘origin’).
#### Sabendo disso, existe uma correlação entre a aceleração e o consumo? De que tipo? A correlação é a mesma para os diferentes países de origem?

```python
plt.figure(figsize=(15,5))
plt.title('Relação entre aceleração e o consumo de combustível pelo país')
sns.scatterplot(x="acceleration", y="mpg", hue = 'origin', data=mpg)
plt.show()
```

```python
#plt.figure(figsize=(15,5))
#plt.title('Relação entre aceleração e o consumo de combustível pelo país')
sns.relplot(y="mpg", x="acceleration", col = 'origin', data=mpg)
#plt.show()
```

Olhano de forma geral, vê-se correlação positiva baixa.
Visualizando por pais já observa que o comportamento no Japão é concentrado em uma faixa, na europa mais espalhada, não demonstrando correlação a primeira vista, já nos EUA os dados são maiores e observa-se a correlação positiva mais clara.

Visualizando tendência pela reta de regressão separadamente.

```python
sns.lmplot(x='acceleration', y='mpg', hue='origin', data=mpg)
```

- Visualizando a reta e a sombra percebe-se que o país EUA demonstra correlação positiva mais clara. Demais países possuem dados espalhados tendo reta positiva porém a sombra se comporta diferente dos EUA.

#### Gera uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível visualizar o ano do modelo (model_year).
#### Existe alguma relação entre as três variáveis?

```python
plt.figure(figsize=(15,5))
plt.title('Relação entre aceleração, consumo de combustível e modelo do veículo')
sns.scatterplot(x='acceleration', y='mpg', hue='model_year', data=mpg)
plt.show()
```

- No visual acima observa-se que os modelos de 70 a 76 possuem menor consumo geral a medida que valor de aceleração aumenta.
- Observa-se o inverso nos modelos com tons mais escuros, que aumentam o consumo no decorrer do eixo x.

```python
sns.lmplot(x='acceleration', y='mpg', hue='model_year', data=mpg)
```

```python
# Dê dois cliques para aumentar o gráfico.
sns.relplot(x="acceleration", y="mpg", col = 'model_year', data=mpg)
```

- Observa-se o consumo por modelo varia muito com base na aceleração, até o modelo 76 não passava de 35, demonstrando um relação, talvez pela tecnologia do período.
-  No conjunto de dados alguns modelos possuem correlação positiva observado pela de regressão e nos visuais em grid. Observa-se que dois modelos a reta foi correlação negativa baixa, porém sem precisão de apontar qual seja pela tonalidade da cor.
- A forma de visualizar com muitas variáveis como foi o caso de modelos de carros não é agradável, muitos dados e cores não dando precisão.
